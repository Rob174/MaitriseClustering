```python
import plotly.express as px
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from h5py import File
from pathlib import Path
%matplotlib inline
```

```python
dataset = {}
with File(Path("../../data/tsplib_normalized.h5").resolve(), "r") as cache:
    for k, v in cache.items():
        dataset[k] = np.copy(v)
```

Distribution du nombre de points

```python
nb_pts = [len(v) for v in dataset.values()]
bins = 10**(np.arange(1,4.1,0.05))
plt.figure(figsize=(8,5))
plt.hist(nb_pts,bins=bins)
plt.xscale('log')
plt.grid(which="both")
plt.xlabel("Number of points")
plt.ylabel("Number of instances")
plt.title("Distribution of the number of points in the dataset TSPLIB")
mean = np.mean(nb_pts)
med = np.median(nb_pts)
plt.axvline(mean, color='r', linestyle='dashed', linewidth=1)
plt.axvline(med, color='g', linestyle='dashed', linewidth=1)
_, max_ylim = plt.ylim()
plt.text(mean*1.1, max_ylim*0.9, 'Mean: {:.2f}'.format(mean))
plt.text(med*1.1, max_ylim*0.9, 'Median: {:.2f}'.format(med))
plt.show()
```
